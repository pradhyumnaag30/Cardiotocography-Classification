```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
df = pd.read_csv("Dataset/Dataset.csv")
df.head()
```

```python
# Dropping useless columns/row

df = df.drop([0], axis=0)
df = df.drop(["FileName","Date", "SegFile"], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
targets = ["NSP", "CLASS"]
for t in targets:
    if t in df.columns:
        print(f"\n{t} value counts:")
        print(df[t].value_counts())
```

```python
sns.countplot(x=df["NSP"])
plt.title("3-Class Target Distribution (NSP)")
plt.show()
```

```python
sns.countplot(x=df["CLASS"])
plt.title("10-Class Target Distribution (CLASS)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(18, 12))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
corr_nsp = corr["NSP"].sort_values(ascending=False)
corr_nsp
```

```python
corr_class = corr["CLASS"].sort_values(ascending=False)
corr_class
```

```python
# Save Cleaned Dataset

df.to_csv('Dataset/Cleaned.csv', index=False) 
```
